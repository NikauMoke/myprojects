```python
df2 = df.drop(["fuel","seller_type"], axis='columns')
df2
```

```python
df["name"].unique()
```

##### **Creating dummy variables for car model**

```python
dummies = pd.get_dummies(df.name)
dummies
```

```python
df3 = pd.concat([df2,dummies],axis='columns')
```

**Now, Drop one column to avoid dummy variable trap**

```python
df4 = df3.drop(['year','name','Volvo XC60 D5 Inscription'], axis='columns')
df4
```

#### **Spliting into X and Y and then using train_test_split**

```python
X = df4.drop(['selling_price'],axis='columns')
y = df4.selling_price
```

```python
X
```

```python
y
```

#### **Splitting into Train set and Test set**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

#### **Model Formation**

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(X_train,y_train)
lr_model.score(X_test,y_test)
```

#### **Lasso Regression** 

```python
from sklearn import linear_model
lasso_model = linear_model.Lasso(alpha=2, selection='random')
lasso_model.fit(X_train,y_train)
lasso_model.score(X_test,y_test)
```

Observation: In this case Lasso regression model gives better accuracy than Linear regression model

```python
def predict_price(name,km_driven,is_automatic,owner,car_age):    
    loc_index = np.where(X.columns==name)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = km_driven
    x[1] = is_automatic
    x[2] = owner
    x[3] = car_age
    if loc_index >= 0:
        x[loc_index] = 1

    return lasso_model.predict([x])[0]
```

#### **Predict price** 

Sample

- Model: Maruti Wagon R LXI Minor
- km_driven:50000
- Transmision: Manual
- Owner: First Owner
- Car_age: 15

```python
predict_price('Maruti Wagon R LXI Minor',50000,0,1,15)
```

- Predicted Price: 3677662
- Expected Price:135000

#### **Evaluating the Model** 

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python
y_pred = lr_model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 =  r2_score(y_test,y_pred)

print(f'Mean absolute error: {mae:.2f}')
print(f'Mean squared error: {mse:.2f}')
print(f'Root mean squared error: {rmse:.2f}')
print(f'R^2 error: {r2:.2f}')
```

```python
print('Training score ',lr_model.score(X_train, y_train))
print('Testing score ',lr_model.score(X_test, y_test))
```
